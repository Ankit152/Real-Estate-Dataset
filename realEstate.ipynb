## Feature Engineering

### Handling Missing Malues

```python
## filled the missing values with the mean of that column
data['RM']=data['RM'].fillna(data['RM'].mean())
```

```python
#checking the changes
data.isna().sum()
```

**There are no missing values left in the dataset**

### Visualisation

```python
plt.figure(figsize=(10,10))
sns.heatmap(data.corr(),vmin=-1,vmax=1)
plt.title("Correlation of the Data")
plt.show()
```

```python
y=data['MEDV']
# checking for number of unique values per column
col_dummy=[]
col_std=[]
data=data.drop('MEDV',axis=1)
for col in list(data.columns):
    n=len(pd.unique(data[col]))
    print("Number of unique elements in the column: ",col," is:   ",n)
    if(n<5):
        col_dummy.append(col)
    else:
        col_std.append(col)
        
print("<<<==================================================>>>")
print("Columns for creating dummies: ",len(col_dummy))
print("Columns for standardisation: ",len(col_std))
```

**There is only one feature which is categorical**

```python
# creating dummy variables
data=pd.get_dummies(data,columns=col_dummy)
data.head()
```

```python
## standardising the dataset
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
data[col_std]=sc.fit_transform(data[col_std])
data.head()
```

```python
from sklearn.model_selection import train_test_split as tts
xtrain,xtest,ytrain,ytest=tts(data,y,test_size=0.2,random_state=20)
print(xtrain.shape,ytrain.shape)
print(xtest.shape,ytest.shape)
```

## Making the Model

```python
from sklearn.ensemble import RandomForestRegressor
m=RandomForestRegressor()
m.fit(xtrain,ytrain)
```

```python
print("Model R^2 score is: ",m.score(xtest,ytest))
```

```python
# without bootstrap
m2=RandomForestRegressor(bootstrap=False)
m2.fit(xtrain,ytrain)
print("Model R^2 score is: ",m2.score(xtest,ytest))
```
